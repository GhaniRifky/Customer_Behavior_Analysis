```python
import pandas as pd

df = pd.read_csv('/content/customer_shopping_behavior.csv', sep=',')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
# cek missing value
df.isnull().sum()
```

```python
# handle nilai null dengan median
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# cek missing value
df.isnull().sum()
```

```python
#rename nama kolom menjadi snakecase (ex. customer_id)
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# cek nama kolom
df.columns
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
# cek kolom age_group
df[['age', 'age_group']].head(10)
```

```python
# buat kolom purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
# cek kemiripan kolom
(df['discount_applied'] == df['promo_code_used']).all()

## hasilnya True, artinya kedua kolom memiliki value yang sama
```

```python
# hapus kolom promo_code_used
df = df.drop('promo_code_used', axis=1)
```
